```python
import subprocess
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_plot(min_threats, max_threats, size, filename):
    subprocess.call(["./main", str(min_threats), str(max_threats), str(size), filename])
    data = np.loadtxt(filename)
    
    plt.figure(figsize=(15, 8))
    plt.plot(np.arange(min_threats, max_threats + 1), data)
    plt.xlabel("Количество потоков", fontsize=16)
    plt.ylabel("Время", fontsize=16)
```

```python
get_plot(1, 32, 20000000, "file.txt")
```
